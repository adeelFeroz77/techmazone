Filtering after GROUPBY can be done by "Having"

```python

q = """
    Select Distinct(Major_category),avg(ShareWomen) as AVG
    from Recent_grads
    group by Major_category
    having AVG > 0.5
"""
pd.read_sql_query(q,conn)
```

```python
q = """
    Select ShareWomen, women, total, (women/total)*100 as WomenPercent from recent_grads
"""
pd.read_sql_query(q,conn)
```

```python
q = """
    Select ShareWomen, women, total, (women/cast(total as float))*100 as WomenPercent from recent_grads
"""
pd.read_sql_query(q,conn)
```

```python
q = """
    Select ShareWomen, women, total, round((women/cast(total as float))*100,2) as WomenPercent from recent_grads
"""
pd.read_sql_query(q,conn)
```

```python
q = """
    Select Distinct(Major), avg(ShareWomen) as AVG
    From Recent_grads
    Group By Major
    having AVG > 0.5
    ORDER BY AVG DESC
    LIMIT 10
"""
pd.read_sql_query(q,conn)
```

```python
q = """
    Select ShareWomen, men, total, round((men/cast(total as float))*100,2) as MenPercent from recent_grads
"""
pd.read_sql_query(q,conn)
```
