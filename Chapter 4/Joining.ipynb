```python
import pandas as pd
import numpy as np
import sqlite3 as sql
```

```python
db = "factbook.db"
conn = sql.connect(db)
```

```python
def read(q):
    return pd.read_sql_query(q,conn)
```

```python
q = """
    Select * from SQLITE_MASTER
    where type = "table"
"""
read(q)
```

```python
facts = read("SElect * from facts")
cities = read("SElect * from cities")
```

```python
q = """
    Select f.id, c.facts_id, f.name as Country, c.name as City,f.population as Country_Population, c.population as City_Population
    from facts as f
    inner join cities as c
    on f.id == c.facts_id
    Order by f.id
"""
read(q)
```

```python
q = """
    Select f.id, f.name as Country,f.population as Country_Population, c.population as City_Population, cast(sum(c.population) as float)/f.population * 100 as Perc
    from facts as f
    inner join cities as c
    on f.id == c.facts_id
    group by Country
    Order by f.id
"""
read(q)
```

```python
q = """
    Select f.id, c.facts_id, f.name as Country, c.name as City,f.population as Country_Population, c.population as City_Population
    from facts as f
    left join cities as c
    on f.id == c.facts_id
    Order by f.id
"""
read(q)
```

```python
q = """
    Select f.id, c.facts_id, f.name as Country, c.name as City,f.population as Country_Population, c.population as City_Population
    from facts as f
    right join cities as c
    on f.id == c.facts_id
    where f.name = 'Pakistan'
    Order by f.id
"""
read(q)
```

```python
q = """
    Select f.id, c.facts_id, f.name as Country, c.name as City,f.population as Country_Population, c.population as City_Population
    from facts as f
    inner join cities as c
    on f.id == c.facts_id
    where f.name = 'Pakistan'
    Order by f.id
"""
read(q)
```

```python
q = """
    Select facts_id , sum(population) as urban_pop from cities
    group by facts_id
"""
read(q)
```

```python
q = """
    Select f.id, f.name as Country,f.population as Country_Population, cast(sum(a.urban_pop) as float)/f.population * 100 as Perc
    from facts as f
    inner join (Select facts_id , sum(population) as urban_pop from cities
    group by facts_id) as a
    on f.id == a.facts_id
    group by Country
    Order by f.id
"""
read(q)
```
