```python
#Invoice Id, Track Id, Track Name, MediaNAme , Quantity, UnitPrice

q = """
    Select IL.Invoice_id, T.track_id, T.name as TrackName, IL.quantity,M.name as MediaName,IL.Unit_Price from invoice_line as IL
    inner join track as T
    on T.track_id = IL.track_id
    inner join Media_type as M
    on M.media_type_id = T.media_type_id
    where IL.Invoice_id = 1
"""
read(q)
```

```python
#Invoice Id, Track Id, Track Name, MediaNAme , Quantity, UnitPrice

q = """
    Select IL.Invoice_id, T.track_id, T.name as TrackName, IL.quantity,M.name as MediaName,IL.Unit_Price,Ar.name as ArtistName from invoice_line as IL
    inner join track as T
    on T.track_id = IL.track_id
    inner join Media_type as M
    on M.media_type_id = T.media_type_id
    inner join album as A
    on A.album_id = T.album_id
    inner join Artist as Ar
    on Ar.artist_id = A.artist_id
    where IL.Invoice_id = 1
"""
read(q)
```

```python
#album title artist name Sales of selected Tract

q = """
    Select A.title, Ar.name, sum(IL.Quantity) as sales from Album as A
    inner join artist as Ar
    on Ar.artist_id = A.artist_id
    inner join track as T
    on T.album_id = A.album_id
    inner join invoice_line as IL
    on IL.track_id = T.track_id
    group by A.album_id
    order by sales DESC
"""

read(q)
```

```python
q = """
    with playlist_info as 
    (Select PL.playlist_id,PL.name,count(PT.track_id) Number_of_Tracks,sum(cast(T.milliseconds as float)/1000) Total_Duration_in_seconds from playlist as PL
    inner join playlist_track as PT
    on PT.playlist_id == PL.playlist_id
    inner join track as T
    on T.track_id == PT.track_id
    group by PL.playlist_id)
    
    Select * from playlist_info
"""
read(q)
```

```python
q = """
    create view playlist_info as 
    Select PL.playlist_id,PL.name,count(PT.track_id) Number_of_Tracks,sum(cast(T.milliseconds as float)/1000) Total_Duration_in_seconds from playlist as PL
    inner join playlist_track as PT
    on PT.playlist_id == PL.playlist_id
    inner join track as T
    on T.track_id == PT.track_id
    group by PL.playlist_id
"""
conn.execute(q)
```
