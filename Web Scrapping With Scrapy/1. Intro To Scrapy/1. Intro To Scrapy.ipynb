```python
import scrapy
import pandas as pd
from scrapy.crawler import CrawlerProcess

class QuotesScrapper(scrapy.Spider):
    name = "Quotes"
    start_urls = ["http://quotes.toscrape.com/"]
    
    
    def parse(self,response):
        quotes = response.css("span.text ::text").getall()
        authors = response.css("small.author ::text").getall()
        
        quoteList = []
        for i in range(len(quotes)):
            item = {
                "author" : authors[i],
                "quotes" : quotes[i]
            }
            quoteList.append(item)
        pd.DataFrame(quoteList).to_csv("Quotes.csv",index=False)
#         print(response.text)

process = CrawlerProcess()
process.crawl(QuotesScrapper)
process.start()
```
