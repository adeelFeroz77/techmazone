```python
import csv
openFile = open("epl_new_dataset.csv",encoding = 'latin-1')
readFile = csv.reader(openFile)
dataset = list(readFile)
```

```python
header = dataset[0]
eplData = dataset[1:]
```

```python
eplData17 = []
```

```python
for i in eplData:
    season  = i[0]
    if season == "2007-08":
        eplData17.append(i)
```

```python
len(eplData17)
```

```python
def matchesPlayed(teamName,dataset):
    count = 0
    for i in dataset:
        if i[2]==teamName or i[3]==teamName:
            count+=1
    return count
```

```python
print(matchesPlayed("Man United",eplData17))
```

```python
teams = []
for i in eplData17:
    team = i[2]
    if not team in teams:
        teams.append(team)
```

```python
teams = sorted(teams)
```

```python
len(teams)
```

```python
table = []
table.append(["Club","MP","W","D","L","GF","GA","GD","Pts"])
```

```python
teams
```

```python
for i in teams:
    matchesPlayed = 0
    goalsScored = 0
    goalsConceded = 0
    wins = 0
    draws = 0
    loses = 0
    points = 0
    for j in eplData17:
        if j[2] == i:
            matchesPlayed +=1
            status = j[6]
            if status == 'H':
                wins+=1
            elif status == 'D':
                draws+=1
            else:
                loses +=1
            goalsScored+=int(j[4])
            goalsConceded+=int(j[5])
        elif j[3] == i:
            matchesPlayed +=1
            status = j[6]
            if status == 'H':
                loses+=1
            elif status == 'D':
                draws+=1
            else:
                wins +=1
            goalsScored+=int(j[5])
            goalsConceded+=int(j[4])
    points = (wins*3)+draws
    table.append([i,matchesPlayed,wins,draws,loses,goalsScored,goalsConceded,abs(goalsScored-goalsConceded),points])
            
            
        
```

```python
len(table)
```

```python
def pointsSort(row):
    return int(row[-1])
```

```python
table = sorted(table[1:], key=pointsSort, reverse = True)
```
