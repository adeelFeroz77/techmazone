```python
def sort(table):
    for i in range(0,len(table)):
        for j in range(i+1,len(table)):
            if table[i][-1]<table[j][-1]:
                table[i],table[j] = table[j],table[i]
            elif table[i][-1] == table[j][-1]:
                if table[i][-2]<table[j][-2]:
                    table[i],table[j] = table[j],table[i]
    return table
```

```python
def solve(season,dataset):
    from openpyxl import load_workbook
    print(f"\t\t\t{season}")
    import pandas as pd
    eplData17 = []
    for i in dataset:
        currentSeason  = i[-1]
        if currentSeason == season:
            eplData17.append(i)
    teams = []
    for i in eplData17:
        team = i[2]
        if not team in teams:
            teams.append(team)
    teams = sorted(teams)
    table = []
    # table.append(["Club","MP","W","D","L","GF","GA","GD","Pts"])
    #ITERATING TEAMS
    for i in teams:
        matchesPlayed =goalsScored =goalsConceded =wins =draws =loses =points = 0
        #ITERATING THROUGH SEASON DATASET
        for j in eplData17:
            # IF TEAM IS AT HOME
            if j[2] == i:
                matchesPlayed +=1
                status = j[6]
                if status == 'H':
                    wins+=1
                elif status == 'D':
                    draws+=1
                else:
                    loses +=1
                goalsScored+=int(j[4])
                goalsConceded+=int(j[5])
            # IF TEAM IS AT AWAY
            elif j[3] == i:
                matchesPlayed +=1
                status = j[6]
                if status == 'H':
                    loses+=1
                elif status == 'D':
                    draws+=1
                else:
                    wins +=1
                goalsScored+=int(j[5])
                goalsConceded+=int(j[4])
        #POINTS FORMULA
        points = (wins*3)+draws
        gd = goalsScored-goalsConceded
        table.append([i,matchesPlayed,wins,draws,loses,goalsScored,goalsConceded,gd,points])
    table = sort(table)
    tableHeader = ["Club","MP","W","D","L","GF","GA","GD","Pts"]
#     print(table)
    df = pd.DataFrame(table,columns=tableHeader)
#     print(df)
    df.name = season
    winTeam = table[0][0]
    print(df)
    print(f"\n\tWinner for Season {season} is {winTeam}")
#     indexx = season[3]
    df.to_excel("Result.xlsx",index = False)
    return winTeam
```

```python
solve("2011-12",eplData)
```

```python
winnerFreq = {}
for i in range(2010,2018):
    win = solve(f"{i}-{(i%100)+1}",eplData)
    if win in winnerFreq:
        winnerFreq[win] += 1
    else:
        winnerFreq[win] = 1
    print("\n\n")
```
