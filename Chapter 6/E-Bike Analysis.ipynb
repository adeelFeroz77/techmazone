```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("day.csv", parse_dates=['dteday'])
data
```

```python
data.info()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['dteday'],data['temp'])
plt.title("Temperature over the period of Months/Quarters")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['dteday'],data['cnt'])
plt.title("Sales over the period of Quarters")
plt.xlabel("Date")
plt.ylabel("Sales Count")
plt.show()
```
