```python
vendor = ""
name = ""
account = ""
vendorList = []
nameList = []
accountList = []
for i in range(0,len(df["Vendor"])):
    currentVendor = df.iloc[i,0]
    currentName = df.iloc[i,1]
    currentAccount = df.iloc[i,2]
    if pd.notnull(currentVendor):
        vendor = currentVendor
        name = currentName
        account = currentAccount
    vendorList.append(vendor)
    nameList.append(name)
    accountList.append(account)
```

```python
df["Vendor"] = vendorList
df["Name"] = nameList
df["St"] = accountList
```
