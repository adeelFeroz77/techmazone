```python
import pandas as pd
```

```python
data = pd.read_excel("Data Comparision.xlsx",'Climax')
```

```python
oldRecords = data.iloc[:41,:2]
newRecords = data.iloc[:48,4:]
```

```python
oldRecords.columns = oldRecords.iloc[0]
newRecords.columns = newRecords.iloc[0]
```

```python
oldRecords = oldRecords[1:]
newRecords = newRecords[1:]
```

```python
oldRecords["Client Name"] = oldRecords["Client Name"].str.replace("\(EXPORT\)","")
oldRecords["Client Name"] = oldRecords["Client Name"].str.strip()
oldRecords
```

```python
newRecords["Client Name"] = newRecords['Client Name'].str.replace("\(EXPORT\)","")
newRecords["Client Name"] = newRecords['Client Name'].str.strip()
newRecords
```

```python
mergedDF = pd.merge(oldRecords,newRecords,how = "outer",on = "Client Name",suffixes=("_Old","_New"))
```

```python
mergedDF["Nominal"] = ""
```

```python
for i in range(len(mergedDF)):
    if str(mergedDF.iloc[i,1])=="nan":
        mergedDF.iloc[i,3] = "Missing Old"
        continue
    if str(mergedDF.iloc[i,2]) == "nan":
        mergedDF.iloc[i,3] = "Missing New"
        continue
    if mergedDF.iloc[i,1] == mergedDF.iloc[i,2]:
        mergedDF.iloc[i,3] = "Matched"
        continue
    mergedDF.iloc[i,3] = "Not Matched"
```

```python
mergedDF = mergedDF.sort_values("Client Name").reset_index(drop=True)
mergedDF
```

```python
mergedDF.to_excel("Combined DataFrame.xlsx",index=False)
```
