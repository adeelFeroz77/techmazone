```python
AllMovies = [["Rank","Movie Name","Movie Link","Movie Year","Movie TMeter","Movie Synopsis","Movie Cast","Movie Director"]]
URLs = 'https://editorial.rottentomatoes.com/guide/essential-movies-to-watch-now/'
count = 1
for i in range(1,3):
    response = requests.get(URLs+f"{i}/")
    content = response.text
    soup = BeautifulSoup(content,'html.parser')
    allData = soup.find_all("div",{"class":"row countdown-item"})
    for movieData in allData:
        movieHeader = movieData.find_all('h2')[0]
        movieName = movieHeader.find('a').text
        movieLink = movieHeader.find('a')['href']
        movieYear = movieHeader.find('span').text
        movieTMeter = movieHeader.find('span',{'class':'tMeterScore'}).text
        movieSynopsis = movieData.find('div',{'class':'info synopsis'}).text
        movieCast = [i.text for i in movieData.find('div',{'class':'info cast'}).find_all('a')]
        movieCast = str(movieCast)[1:-1]
        movieDirector = [i.text for i in movieData.find('div',{'class':'info director'}).find_all('a')]
        movieDirector = str(movieDirector)[1:-1]
        
        movieDetails = [count,movieName,movieLink,movieYear,movieTMeter,movieSynopsis,movieCast,movieDirector]
        AllMovies.append(movieDetails)
        count+=1

print(AllMovies)
```

```python
df = pd.DataFrame(AllMovies[1:],columns=AllMovies[0])
df
```
